```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("breast_cancer_enhanced_dataset.csv")
```

```python
print(df.head())
```

```python
df['concavity_density']
```

```python
df.describe()
```

```python
df=df.drop('id',axis=1)
```

```python
df.head()
```

```python
df['diagnosis']=df['diagnosis'].map({"B":0,"M":1})
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
mx=MinMaxScaler()
x_train = mx.fit_transform(x_train)
x_test = mx.transform(x_test)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rnd = RandomForestClassifier(n_estimators=100,random_state=42)
rnd.fit(x_train,y_train)
y_pred = rnd.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
# Train data ki accuracy
train_pred = rnd.predict(x_train)
train_acc = accuracy_score(y_train, train_pred)

# Test data ki accuracy
test_pred = rnd.predict(x_test)
test_acc = accuracy_score(y_test, test_pred)

print(f"Training Accuracy: {train_acc * 100:.2f}%")
print(f"Testing Accuracy: {test_acc * 100:.2f}%")
```

```python
import joblib

# Apne trained random forest model ko save karein
joblib.dump(rnd, 'random_forest_model.joblib')
print("Model locally save ho gaya hai!")
```
